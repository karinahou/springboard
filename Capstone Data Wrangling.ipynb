```python
# Turn the string of the date into a datetime object to extract the month
def str_to_datetime(date_str):
    return datetime.strptime(date_str, '%Y-%m-%d').month

#print each date string into the number month in a new column
month_col = df_calendar['date'].apply(str_to_datetime)
```

```python
#turns the month number into the name of the month
def num_to_month(num):
    if num == 1:
        return "Jan"
    if num == 2:
        return "Feb"
    if num == 3:
        return "Mar"
    if num == 4:
        return "Apr"
    if num == 5:
        return "May"
    if num == 6:
        return "Jun"
    if num == 7:
        return "Jul"
    if num == 8:
        return "Aug"
    if num == 9:
        return "Sep"
    if num == 10:
        return "Oct"
    if num == 11:
        return "Nov"
    if num == 12:
        return "Dec"
```

```python
#made a new single column containing the name of each month for each value of df_calendar
month_new_col = month_col.apply(num_to_month)
```

```python
#added the month name to the calendar
df_calendar_month = pd.concat([df_calendar, month_new_col], axis=1)
```
